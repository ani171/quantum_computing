```python
from qiskit import *
```

```python
circuit=QuantumCircuit(2,2)
# or
# q_reg=QuantumRegister(2)
#c_reg=ClassicalRegister(2)
#circuit=QuantumCircuit(q_reg,c_Reg)
```

```python
circuit.draw()
```

```python
%matplotlib inline
```

```python
circuit.draw(output='mpl')
```

```python
circuit.h(0) #Applying Hedamard gate to q0
circuit.draw(output='mpl')
```

```python
circuit.cx(0,1) # q0 is control qbit and q1 is target qbit
circuit.draw(output='mpl')
```

```python
circuit.measure([0,1],[0,1]) #measure q0 and q1 and associate them with classic bit 0 and 1
circuit.draw(output='mpl')
```

```python
simulator=Aer.get_backend('qasm_simulator') #Run this command pip install qiskit-aer before executing the simulator command 
```

```python
execute(circuit,backend=simulator)
```

```python
result=execute(circuit,backend=simulator).result()
```

```python
from qiskit.visualization import plot_histogram
```

```python
plot_histogram(result.get_counts(circuit))
```
